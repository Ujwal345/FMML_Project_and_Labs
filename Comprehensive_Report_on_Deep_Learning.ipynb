**INTRODUCTION**

Deep learning, a specialized field within machine learning, revolves around neural networks capable of learning intricate representations. This approach emphasizes training models to discern patterns and make decisions autonomously, minimizing reliance on human intervention. Notably, deep learning has excelled in domains like image recognition, speech processing, and natural language understanding by leveraging its capacity to glean insights from extensive datasets.

**FUNDAMENTALS OF DEEP LEARNING**

**NEURAL NETWORK**

Neural Networks: Deep learning is based on artificial neural networks, which
are inspired by the biological neurons in the human brain. A neural network consists of layers of interconnected nodes (neurons) that process and transform input data into meaningful output.

1.	Layers: Neurons are organized into layers in a neural network. There are typically three types of layers:
•	Input Layer: Receives input data (e.g., images, text).
•	Hidden Layers: Intermediate layers between the input and output layers where computations occur. Deep learning models have multiple hidden layers, allowing them to learn complex representations.
•	Output Layer: Produces the final output of the model (e.g., classification scores, predicted values).
2.	Neurons: These are fundamental units in a neural network, analogous to biological neurons. Each neuron receives inputs, performs a computation (usually a weighted sum of inputs plus a bias), and applies an activation function to produce an output.

Activation functions: Activation functions are crucial components in neural networks, introducing non-linearity to enable the learning of complex patterns from data. Here are commonly used activation functions:
 Sigmoid Function:
•	Range: Squashes input values to a range between 0 and 1.
•	Formula: σ(x)=11+e−x\sigma(x) = \frac{1}{1 + e^{-x}}σ(x)=1+e−x1
•	Use: Historically used in output layers for binary classification tasks where the output represents probabilities.

 Tanh Function:
•	Range: Squashes input values to a range between -1 and 1.
•	Formula: tanh(x)=ex−e−xex+e−x\text{tanh}(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}}tanh(x)=ex+e−xex−e−x
•	Use: Similar to sigmoid but centered around zero, often used in hidden layers to introduce non-linearity.
 ReLU (Rectified Linear Unit):
•	Definition: Outputs the input directly if it is positive; otherwise, outputs zero.
•	Formula: ReLU(x)=max⁡(0,x)\text{ReLU}(x) = \max(0, x)ReLU(x)=max(0,x)
•	Use: Widely used in hidden layers due to its simplicity and effectiveness in accelerating convergence during training. It helps alleviate the vanishing gradient problem encountered with sigmoid and tanh activations.
Training Process:
1.	Labeled Data:
o	Neural networks are trained using labeled datasets where each data point has a known target or label (e.g., image category, numerical value).
2.	Loss Function:
o	Measures the difference between the model's predictions and the actual labels.
o	Examples include:
	Cross-Entropy Loss: Used for classification tasks.
	Mean Squared Error (MSE): Used for regression tasks.
3.	Backpropagation:
o	Calculates gradients of the loss function with respect to each network parameter.
o	Gradients are used to update weights and biases, adjusting them to minimize the loss.
Optimization Algorithms:
1.	Gradient Descent:
•	Stochastic Gradient Descent (SGD): Updates weights iteratively based on gradients computed from random mini-batches of data.
2.	Advanced Optimizers:
•	Adam: Combines momentum and RMSprop techniques to adjust learning rates adaptively for each parameter.
•	RMSprop: Adjusts learning rates based on the magnitudes of recent gradients for each parameter.

**Convolutional Neural Networks (CNNs)**

Convolutional layers apply filters (kernels) to the input image and create feature maps that highlight various features such as edges, textures and patterns.


a. Filters and Character Maps
- Filters: small tables that slide over the input image to identify specific features. Each filter creates a feature map that represents the presence of the filter feature at different points in the input.
- Pitch and Padding: The pitch determines how much the filter moves at each step. Padding involves adding edges to the input image to control the spatial dimensions of the output feature maps.

b. Activation functions in CNNs
Convolutional layers are usually followed by activation functions (eg ReLU) to increase nonlinearity.

Pooling Layers:
Reduce the spatial dimensions of the feature maps while retaining important information, typically using operations like max pooling.
Types of Pooling
Max Pooling: Extracts the maximum value from each window of the input feature map.
Average Pooling: Computes the average value of each window of the input feature map.
Global Average Pooling: Computes the average value of the entire feature map.
Global Max Pooling: Computes the maximum value of the entire feature map.
Fully connected layers:
Fully connected layers in convolutional neural networks (CNNs) play a crucial role in integrating extracted features for final decision-making.
•	Operation: Flatten the feature maps extracted by convolutional and pooling layers into a vector format.
•	Purpose: Connect each element of the vector to every neuron in a subsequent layer, enabling the network to learn complex relationships between features.
•	Application: Typically used in the final layers of CNNs for tasks such as image classification or regression, where the network outputs predictions based on the learned representations.

**RECURRENT NEURAL NETWORKS (RNNs)**

RNNs are tailored to process sequential data by retaining memory of previous inputs, making them adept for tasks such as time series forecasting and natural language processing. Their architecture includes recurrent connections that enable information persistence over time, facilitating effective learning from sequences in various applications.

1. Basic RNN
In a standard RNN, the output of the previous time step is fed back to the network as the input of the current time step, allowing the network to retain memory of previous inputs.

a. Structure
- Hidden state: The hidden state is updated at each time step based on the current input and the previous hidden state.
- Output: The output at each time step is based on the current hidden state.

b Limitations
Basic RNNs suffer from the vanishing gradient problem, where the gradients become very small, making it difficult to learn long-term dependencies.

Long Term Short Memory (LSTM):
Long Short-Term Memory (LSTM) networks are a type of recurrent neural network (RNN) architecture designed to overcome the limitations of traditional RNNs in learning long-term dependencies.
 Cell State: Acts as a conveyor belt running through the entire chain of LSTM units, allowing information to flow along the sequence without alteration.
 Forget Gate: Decides what information to discard from the cell state.
 Input Gate: Updates the cell state with new information.
 Output Gate: Determines the output based on the updated cell state.

Bidirectional RNN:
Bidirectional Recurrent Neural Networks (RNNs) are a variant of RNNs that enhance the learning capability by processing input sequences in both forward and backward directions simultaneously.

Two Directions: Consist of two separate RNNs: one processing the sequence from the beginning to end (forward RNN), and the other from end to beginning (backward RNN).
Concatenation: Outputs from both directions are typically concatenated or combined before being passed to the next layer or task.

**TRANSFORMERS**

Transformers are an advanced architecture designed for processing sequential data without relying on recurrent connections. They employ self-attention mechanisms to assess the significance of various elements within the input sequence, thus optimizing their performance in tasks such as natural language processing and image recognition.

Attention Mechanism:
The attention mechanism enables models to selectively focus on relevant parts of the input sequence when making predictions.
a. Self-Attention
•	Scaled Dot-Product Attention: Computes attention scores by scaling the dot product of query and key vectors, adjusted by the square root of the key vector dimension.
•	Multi-Head Attention: Utilizes multiple attention heads to capture diverse features and dependencies within the input sequence simultaneously.
Applications of Transformers:
Transformers find extensive applications in natural language processing tasks such as machine translation, text summarization, and language modeling. They are also adapted for tasks beyond traditional NLP domains.
a. BERT (Bidirectional Encoder Representations from Transformers):-BERT is a transformer-based model pre-trained on large text corpora, excelling in various NLP tasks by learning deep bidirectional representations of language.
b. GPT (Generative Pretrained Transformer):-GPT is a generative transformer known for tasks such as text generation, completion, and summarization. Versions like GPT-3 and GPT-4 are highly advanced, capable of generating human-like text and performing complex language tasks.
Encoder-decoder architecture
Transformers consist of an encoder that processes an input sequence and a decoder that generates an output sequence. Each layer of the encoder and decoder has an attention mechanism and forward neural networks.

a. Encoder
- Self-Aware Layers: Capture input-sequence dependencies.
- Fed-Forward Layers: Use non-linear transformations to output introspective layers.

b. Decoder
- Mesh self-attention: Prevents the model from paying attention to future positions in the print sequence.

- Encoder and Decoder Beware: Monitors encoder output to add information about the sequence
